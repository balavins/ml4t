```python
import pandas as pd
```

```python
df = pd.read_csv('us_stocks.csv')
```

```python
df['date'] = pd.to_datetime(df['date'], format = "%Y-%m-%d")
```

```python
fmask = df['date'] >= '2000-01-01'
```

```python
filterDf = df.loc[fmask].set_index(['ticker','date'])
```

```python
resultDf = filterDf[['adj_open','adj_high','adj_low','adj_close','adj_volume']]
```

```python
resultDf.columns = resultDf.columns.str.lstrip("adj_")
```

```python
resultDf.to_hdf("cleaned_us_stocks.h5", "df")
```
